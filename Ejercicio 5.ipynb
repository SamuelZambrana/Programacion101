# Ejercicio 5. Tipo de triangulo

## Enunciado.

El principio de desigualdad triangular dice que tres segmentos pueden formar un triangulo si la suma de las longitudes de cualquier par de lados es mayor que la longitud del otro.

Por otro lado los triangulos son:
- Isósceles: si dos lados son iguales y el otro no
- Escaleno: si los tres lados son distintos
- Equilátero: si los tres lados son iguales.

Debes crear una función con tres longitudes como entrada y que imprima el tipo de triangulo que es o que los tres segmentos no pueden formar un triángulo.

Además de verificar el tipo de triángulo o que no lo es, debes llevar la cuenta de cuantos triángulos de cada clase o no triángulos han sido verificados. Hazlo con variables globales.

## Ayuda a la implementación

Dejo un par de ejemplos de tres segmentos que no pueden formar un triángulo (5, 10, 5). 5 + 5 no es mayor que 10 por tanto no puede formar un triángulo. Lo mismo pasa con (5, 11, 4)

Sin embargo (5, 7, 11) sí forman un triángulo valido porque:

- 5 + 7 > 11
- 5 + 11 > 7
- 7 + 11 > 5

La comprobación de isósceles, equilátero y escaleno te la dejo a ti.
