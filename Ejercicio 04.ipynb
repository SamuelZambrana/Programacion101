# Ejercicio 4. Apuestas de la ruleta francesa

## Enunciado

La ruleta europea tiene 36 números (del 1 al 36), más el cero. 

El cero es de color verde, mientras que los múmeros
2, 4, 6, 8, 10, 11, 13, 15, 17, 20, 22, 24, 26, 28, 29, 31, 33 y 35 son rojos
y los 
1, 3, 5, 7, 9, 12, 14, 16, 18, 21,23, 25, 27, 30, 32 y 24 son negros.

Las apuestas sencillas que podemos hacer son:
- a un número específico (esta no se debe construir)
- Par: cualquier número par (no incluye el cero)
- Impar: cualquier número impar (no incluye el cero)
- Rojo: si el número está incluido en el grupo de los rojos
- Negro: si el número está incluido en el grupo de los negros
- Falta: si el número está incluido entre el 1 y el 18 (ambos inclusive)
- Pasa: si el número está incluido entre el 19 y el 36 (ambos inclusive)

Existen dos apuestas más que son las docenas. La ruleta se divide en 3 docenas, 
- 1ª.  1 - 12
- 2ª.  13 - 24
- 3ª.  25 - 36

Pero las docenas solo se pueden elegir siendo par o impar, por ejemplo `primera par` que incluirá entonces los números 2, 4, 6, 8, 10 y 12. La imagen del siguiente [enlace](https://www.casino.es/ruleta/ruleta-francesa/) facilitará entender esta última apuesta.

Debes construir predicados tales que recibiendo como parámetro de entrada un número entre 0 y 36 inique si la apuesta del predicado es ganadora o no.




## Ayuda a la implementación

Al no manejar todavía instrucciones de secuencia de datos, deberás gestionar los grupos de valores que forman rojo o negro con condiciones lógicas complejas
