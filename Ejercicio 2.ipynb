# Ejercicio 2. Interes simple

## Enunciado

Crear una función que calcule el interés simple, sabiendo que se calcula con la siguiente fórmula:

>$ A = P · (1 + rt) $

donde *A* es la cantidad ganada, *P* la cantidad invertida, *r* el interes y *t* los años transcurridos desde el inicio de la inversión

Como en el ejercicio anterior modifica la cantidad inicial añadiéndole el interés calculado. Hazlo de las dos formas vistas en el ejercicio anterior.

**Restricciones**

*r (interés)*: debe informarse como porcentaje, es decir 3% se informa como 3 y o como 0.03 (tanto por 1).


## Ayuda a la implementación

Recuerda que puedes usar el modificador global porque son ejercicios, intenta siempre no utilizarlo. Por eso pedimos las dos soluciones

```python
A = 10000
año = 1
def simple_interest(investment, ratepercent, time):
    rate = ratepercent/100
    return round(investment*(1+rate*time), 2)
```

```python
A = simple_interest(A, 3, 1)

print('pasado {} año haganado tienes {}'.format(año, A))
año = año +1
```
