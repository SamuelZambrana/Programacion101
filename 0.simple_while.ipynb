Crea la función `sum_even` que recibe una lista de números y devuelve la suma de los números pares.

Recuerda los 4 elementos de la iteración.

````python
def sum_even(numberList):
  '''Recibe una lista y devuelve la suma de los numeros pares'''

  sumEven=0
  index=0

  while index<len(numberList):      # La unica condicion de salida es que se llegue al final de la lista
    if numberList[index]%2==0:      # Seleccionamos los pares
      sumEven += numberList[index]  # Añadimos valor a acumulador
    else:                           # En caso de no ser par no hacemos nada
      pass
    index += 1                      # Incremento del index para ver el siguiente elemento
  return  sumEven
````

````python
def starts_with_vowel(inputString):
  '''Predicado que determina si una cadena empieza por cadena'''
  return inputString[0] in 'aeiouAEIOU'

def concat_vowels(stringList):
  '''la funcion revibe una lista de cadenas y devuelve una cadena que concatena las cadenas que empiezan por vocal'''
  concatString=''
  index=0

  while index<len(stringList):                # La unica condicion de salida es que se llegue al final de la lista
    if starts_with_vowel(stringList[index]):  # Seleccionamos los elementos que empiezan por volcal
      concatString += stringList[index]       # Concatenamos a la cadena de salida
    else:                                     # En caso de no ser par no hacemos nada
      pass
    index += 1                                # Incremento del index para ver el siguiente elemento
  return  concatString
````

Crea la función `less_than(numbers, n)` que recibe una lista de números y un número n. Devuelve la lista de aquellos números que son menores o iguales a n.

1. ¿Qué tipo de función es (compresor, selector o transformador)?

````python
def less_than(numbers, n):
  '''recibe una lista de números y un número n. Devuelve la lista de aquellos números que son menores o iguales a n.'''
  lessThanN=[]
  index=0
  while index<len(numbers):                # La unica condicion de salida es que se llegue al final de la lista
    if numbers[index] <=n:                 # Seleccionamos los elementos menores o iguales que n
      lessThanN.append(numbers[index])     # Añadimos a la lista de salida
    else:                                  # En caso de no ser menores o iguales que n
      pass
    index += 1                             # Incremento del index para ver el siguiente elemento
  return  lessThanN
````

Crea la función `add(numbers, n)` que recibe una lista de números y devuelve una nueva lista con los elementos originales + n.

Por ejemplo:

`add([1,2,3,4,5], 2) --> [3, 4, 5,7]`

````python
def add(numbers, n):
  ''' recibe una lista de números y devuelve una nueva lista con los elementos originales + n.'''
  increseList=[]
  index=0
  while index<len(numbers):                # La unica condicion de salida es que se llegue al final de la lista

    increseList.append(numbers[index]+n)   # Añadimos a la lista de salida, el elemento + n

    index += 1                             # Incremento del index para ver el siguiente elemento
  return  increseList
````

````python
def find_one(haystack, needle):
  ''' busca un elemento en una lista y devuelve verdadero o falso'''
  index = 0                       #Inicio de variables, index para recorrer, found para devolver el resultado y repeatCount para almacenar las repeticiones

  found = False
  while not found and index < len(haystack):
    if haystack[index] == needle:
      found = True

    index = index + 1
    return found
````

````python

def find_n(haystack, needle, repeat=1):
  ''' busca una cantidad de repeticiones de elementos en una lista y devuelve verdadero o falso'''

  index = 0                                     #Inicio de variables, index para recorrer, found para devolver el resultado y repeatCount para almacenar las repeticiones
  found = False
  repeatCount =0
  while not found and index < len(haystack):    # Las condiciones de salida son que aun haya sido encontrado las veces deseadas o que lleguemos al final de la lista.
    if haystack[index] == needle:               # En caso de encontrar un elemento buscado aumentamos la chacha en 1
      repeatCount += 1
    if repeatCount == repeat:                   # Si hemos llegado a la chacha objetivo camiamos found a True
      found = True
    index = index + 1
  return found
````

````python

def find_n2(haystack, needle, repeat=1):
  ''' busca una cantidad de repeticiones de elementos en una lista y devuelve verdadero o falso'''

  index = 0                                     #Inicio de variables, index para recorrer, found para devolver el resultado y repeatCount para almacenar las repeticiones
  found = False
  repeatCount =0
  while haystack[index:]:                       # Las condiciones de salida son que aun haya sido encontrado las veces deseadas o que lleguemos al final de la lista.
    if find_one(haystack[index:], needle):      # En caso de encontrar un elemento buscado aumentamos la chacha en 1
      repeatCount += 1
    if repeatCount == repeat:                   # Si hemos llegado a la chacha objetivo camiamos found a True
      found = True
    index = index + 1
  return found
````

````python
def find_n_Fernando(list, needle, n):
    """
    Devuelve True si en list hay n o más ocurrencias de needle
    False si hay menos o si n < 0
    """
    # si n >= 0...
    if n >= 0:
        # Incializamos el índice y el contador
        index = 0
        count = 0

        # mientras no hayamos encontrado al elemento n veces o no hayamos terminado la lista...
        while count < n and index < len(list):
            # si lo encontramos, actualizamos el contador
            if needle == list[index]:
                count = count + 1

            # avanzamos al siguiente elemento
            index = index + 1
        # devolvemos el resultado de comparar contador con n
        return count >= n
    else:
        return False
````
