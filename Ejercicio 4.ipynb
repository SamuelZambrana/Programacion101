# Ejercicio 4. Cuantas veces me has llamado

## Enunciado

Debes crear dos funciones para calcular la edad humana de un perro. En la primera función por cada año de perro se deben contabilizar 7 años, de forma que si el perro tiene un año, el resultado de aplicar la función será 7. 

$edad7(x) = 7x$

La segunda función tiene dos umbrales. Si la edad del perro es menor o igual que dos años se multiplica por 10,5 (a los 2 años un perro es adulto). Y a partir de esa edad cada año real equivale a 4 años humanos.

$edad104(x) = \begin{cases}
10.5x & \text{ si } x \leq 2 \\ 
4x & \text{ si }  x > 2
\end{cases}$

Lo interesante de estas dos funciones es que vamos a contabilizar cuantas veces se ejecutan. Para ello debes crear tres variables. Una variable llamada `contador_lineal` y otra llamada `contador_bifase`, la primera contabilizará cuantas veces se ha llamado a la función del 7 y la segunda cuantas veces se ha llamado la del 10.5. Una tercera variable `contador` llevará la cuenta de ambos.

## Ayuda a la implementación

Solo puedes contar la ejecución de las funciones desde dentro y desde ahí modificar los contadores.
