Crea una función `sum_all` que recibe 3 parámetros:
* Un valor de inicio
* Un valor final
* Un valor de paso

y devuelve la suma de todos los valores de inicio a fin  tomados de paso en paso.

Es decir,

`sum_all(1, 10, 2)` sumaría 1,3,5,7 y 9 y devuelve = 25

Recuerda usar `range`

````python
def sum_all(start, end, step):
  sumAll = 0
  for value in range(start, end+1, step):
    sumAll += value
  return sumAll
````

Crea la función `factorial(n)` que recibe un número y devuelve su factorial.

Recuerda que el factorial (representado con un !) de un número n es la multiplicación de todos los números enteros de 1 hasta n:

$$4! = 4 * 3 * 2 * 1$$

sabiendo que $1! = 1$ y $0! = 1$

Hazlo usando `range` yendo de 1 hasta `n`.

````python
def factorial(n):
  '''Revico un numero y
  devuelvo su factorial
  4!=4∗3∗2∗1
  '''
  #inicializo el acumulador a 1 que es el neutro de la multiplicacion
  fact = 1
  #recorro los valores en el rango 1 a n
  for value in range(1, n+1):
  #acumulo la multiplicacion
    fact =  fact*value
  return fact
````

````python
def is_prime(n):
  '''recibe un numero y devuelve si es primo'''
  #valor indical de retorno y contador de iteraciones para ver eficiencia
  isPrime = False
  iteracion=0
  if n<2:
    #si en numero es menor que 2 no es primo
    isPrime = False

  else:

    #recorro el rango de numeros desde el 2 hasta la mitad del numero
    for factor in range(2, n//2+2):

        if n%factor ==0 and factor !=n:

          isPrime = False
          break

        else:

          isPrime = True
        iteracion +=1


  return isPrime,iteracion
````

````python
def is_primecodellama(n):
    """Check if integer n is a prime number"""
    # make sure n is a positive integer
    n = abs(int(n))

    # 0 and 1 are not primes
    if n < 2:
        return False
    # 2 is the only even prime number
    if n == 2:
        return True
    # all other even numbers are not primes
    if not n & 1:
        return False
    # range starts with 3 and only needs to go up the square root of n for all odd numbers
    for x in range(3, int(n**0.5) + 1, 2):
        if n % x == 0:
            return False

    return True
````

````python
def range_primellama(n):
  primeList =[]
  for number in range (1,n):
    if is_primecodellama(number):
      primeList.append(number)
  return primeList
````

## (6)

Crea la función `rev` que recibe una secuencia (lista o cadena) y la devuelve invertida:

`rev([1,2,3,4]) == [4,3,2,1]`
`rev["Hola"] == "aloH"`

Pista: Se trata de un trasnformador y puedes unar `range` con el tercer parámetro para iterar por los índices hacia atrás.


````python
def implode(stringList):
  # Acumulador con valor inicial
  implodeSring = ''
  # Iteramos
  if type(stringList)!=str:
    for element in stringList:
  # Aplicamos la compresion
     implodeSring = implodeSring + element
  return implodeSring
def explode(myString=''):
  # Acumulador con valor inicial
  explodedList = []
  # Iteramos
  for character in myString:
  # Aplicamos la transformacion requerida
    explodedList.append(character)
  return explodedList
````

````python
def recover_type(preType,data):
  '''recupera el tipo original'''
  if preType is list:
    result = data
  else:
    result = implode(data)
  return result

def rev(iterable):
  '''La funcion recibe una secuencia (lista o cadena)
  y la devuelve invertida'''

  #convierto el elemento a lista para unificar el tratamiento posterior
  toListIterable=explode(iterable)

  #inicio la cadena donde guardare el resultado

  revIterable = []

  #itera de atras adelante para recorrer los elemmentos en orden inverso y meterlos  en la lista resultado

  for item in range(len(toListIterable)-1,-1,-1):
    revIterable.append(toListIterable[item])

  # En caso de haber recivido una lista devuelbo la lista que resultante de la iteración
  # En caso contrario uso implode para volver a convertirlo en una cadena

  result = recover_type(type(iterable),revIterable)

  return result

````
