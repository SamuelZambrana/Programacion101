## Enunciado

Crea una función que calcule la retención anual del sueldo en función de su monto. La retención sigue estos criterios:

1. Si el sueldo es menor o igual a 10000 anuales, la retención es cero
2. Si el sueldo es mayor que 10000 y menor o igual a 15000, la retención será el 2% de la parte del sueldo que exceda los 10000.
3. Si el sueldo es mayor que 15000, la retención será 300 + el 5% de la parte del sueldo que exceda los 15000

$ret(s) = \begin{cases}
0 & \text{ si } s \leq 10000 \\ 
0.02(s - 10000) & \text{ si }  10000 < s \leq 15000\\ 
300 + 0.05(s-15000) & \text{ si } s > 15000 
\end{cases}$

Plantéalo de nuevo para modificar el sueldo tras aplicar la retención con y sin global
