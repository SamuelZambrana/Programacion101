# Ejercicio 3. Es Bisiesto

## Enunciado

Un año es bisiesto en el calendario Gregoriano, si es divisible entre 4 y no divisible entre 100, y también si es divisible entre 400. Crea un predicado (función que devuelve un booleano) que nos diga si un año es o no bisiesto.

```
esBisiesto(1900) -> False
esBisiesto(1972) -> True
esBisiesto(2000) -> True
esBisiesto(1970) -> True
```



## Ayuda a la implementación

Intenta llegar a una solución sin ifs anidados, o mejor que sea una única expresión lógica. Aunque para llegar a ella primero debas usar ifs, elses e ifs anidados.

Con estos ejemplos podrás implementar la solución:

- 1970 NO es bisiesto porque no es divisible entre 4
- 1972 es bisiesto porque es divisible entre 4 y no entre 100
- 1900 NO es bisiesto porque es divisible entre 4 pero también entre 100
- 2000 es bisiesto porque es divisible entre 4, y siendo divisible entre 100 también lo es entre 400 (1900 no es divisible entre 400)
