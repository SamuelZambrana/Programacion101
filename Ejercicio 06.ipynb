# Ejercicio 6. Igualdad en un trío

## Enunciado

Crea una función que devuelva qué numeros son iguales de tres números de entrada (a, b, c). De modo que:
- si no hay ningun número igual a otro devuelva -1
- si son iguales los tres devuelva 1, 2, 3
- y si nos iguales por parejas devuelva las posiciones que son iguales, por ejemplo si son iguales a y c, devolvera 1, 3, a y b, devolverá 1, 2, y así con el resto de condiciones.

Utiliza una única función para ello
