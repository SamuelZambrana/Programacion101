# Ejercicio 10. Sueldo neto mensual

## Enunciado

Debes hacer un conjunto de funciones que te permitan calcular el sueldo neto mensual de una persona teniendo en cuenta lo siguiente:

- Sueldo bruto anual: Te lo informarán en euros
- Retenciones IRPF: Calculadas como en el ejercicio 3. Al calcularse sobre el total anual, se aplicará 1/12 en cada nómina
- Retención de la SS: 9% del sueldo bruto
- Número de horas extra realizadas. Se retiene la SS, pero no IRPF. Se pagan a 15 euros la hora
- El empleado tiene 12 nóminas anuales.

Hacer un programa que pida el bruto anual, y el número de horas extra hechas en el mes. Debe devolver el sueldo neto. 

Estaría bien ver también cuanto retiene hacienda y cuanto a la SS (en el mismo mes).

## Ayuda a la implementación

Utiliza tantas funciones como necesites. No modifiques variables globales dentro de las funciones e intenta evitar el shadowing.
