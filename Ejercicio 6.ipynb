# Ejercicio 6. Decaimiento radiactivo de un elemento

## Enunciado

La fórmula para medir el decaimiento radiactivo (la cantidad de elemento radiactivo que sigue quedando) pasado un tiempo es

$N(t) = N_{0} e ^{(-\lambda t)}$

donde 
- $N_{0}$ = Masa inicial del elemento en gramos
- $t$ = tiempo en días
- $\lambda$ = constante de desintegración en 1/días
- $e$ = número e

Y $N(t)$ es lo que queda realmente del isótopo transcurrido ese tiempo

Se pide crear la función de decaimiento radiactivo y mostrar cuanto cesio-137 quedaría tras 10, 20 y 30 días, si partimos de 1000g de cesio-137. La cantidad de Cs-137 restante debe estar en una única variable, no en tres. 

La constante de desintegración del cesio-137 ($\lambda _ {Cs-137}$) es 0,0871/días


## Ayuda a la implementación

De nuevo puedes hacerlo con la variable cantidad restante como global o no.
