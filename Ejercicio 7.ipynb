# Ejercicio 7

## Enunciado. Determinar si un número es par y múltiplo de 3

Crea una función (o las que necesites) para determinar si un número es a la vez par y múltiplo de tres. Utiliza la evaluación perezosa en la función principal.

Con variables globales (como en el ejercicio 4) demuestra que la evaluación es perezo


## Ayuda a la implementación

Recuerda que la evaluación perezosa puede hacerse por cortocircuito usando or o and, de manera que:
- en el caso de or, si el primer término de la expresión a evaluar es Verdadero el segundo ya no hace falta evaluarlo
- en el caso de and, si el primer término de la expresión es Falso, el segundo ya no hace falta evaluarlo
